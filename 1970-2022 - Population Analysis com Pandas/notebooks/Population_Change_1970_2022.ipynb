```python
# Qual é o país mais populoso de cada continente em 2022? # Usando .groupby(), .loc e .idxmax() para encontrar o país mais populoso de cada continente

# Primeiro, agrupamos os dados por 'Continente' e encontramos o índice do país com a maior população em cada grupo.
idx = df.groupby('Continent')['2022 Population'].idxmax()

# Em seguida, usamos os índices encontrados com .loc para obter as linhas completas do DataFrame.
most_populous = df.loc[idx]

# Exibindo o resultado final
print("País mais populosos por continente em 2022:")
most_populous
```

```python
# Existe alguma correlação entre a área de um país (Area (km²)) e sua população em 2022 (2022 Population)?

# Definir a função get_correlation_strength fornecida anteriormente
def get_correlation_strength(r):
    abs_r = abs(r)

    if 0.9 <= abs_r <= 1.0:
        return "Muito forte"
    elif 0.7 <= abs_r < 0.9:
        return "Forte"
    elif 0.5 <= abs_r < 0.7:
        return "Moderada"
    elif 0.3 <= abs_r < 0.5:
        return "Fraca"
    else:  # abs_r < 0.3
        return "Muito fraca / Nula"

# Calcular a matriz de correlação conforme o código do usuário
print("Calculando a matriz de correlação...")
correlacao = df[['Area (km²)', '2022 Population']].corr()
print("Matriz de Correlação:")
print(correlacao)
print("-" * 30)

# Extrair o valor da correlação entre as duas colunas
# Pegamos o valor específico na intersecção da linha 'Area (km²)' e da coluna '2022 Population'
valor_correlacao = correlacao.loc['Area (km²)', '2022 Population']

print(f"O coeficiente de correlação é: {valor_correlacao:.6f}")
print("-" * 30)

# Usar a função para obter a força da correlação
forca_correlacao = get_correlation_strength(valor_correlacao)

print(f"A força da correlação é: {forca_correlacao}")
```

```python
# Calcule a população mundial total para cada ano disponível no dataset (1970, 1980, 1990, 2000, 2010, 2015, 2020, 2022).

def soma_população(df):
    colunas_population = [col for col in df.columns if col.endswith('Population')]
    lista = []
    
    for i in colunas_population:
        sum = df[i].sum()
        info = [i, sum]
        lista.append(info)
        
    return lista
        

população_mundial_total = soma_população(df)

# Transforma a lista em um DataFrame, especificando os nomes das colunas
df_população = pd.DataFrame(população_mundial_total, columns=['Anos', 'Total_População'])

df_população['Anos'] = df_população['Anos'].str.extract(r'(\d{4})')
df_população = df_população.sort_values(by='Anos', ascending=True).reset_index(drop=True)
df_população
```

```python
# Outliers: Identifique possíveis outliers na coluna Growth Rate. Países com taxas de crescimento muito altas ou muito baixas.
# Dica: Você pode calcular o Z-score ou usar o método do intervalo interquartil (IQR).

# O método do Intervalo Interquartil (IQR) é uma forma robusta de encontrar outliers.
# Calcular Q1 (primeiro quartil), Q3 (terceiro quartil) e IQR

# 1. Análise Descritiva da Coluna 'Growth Rate'
print("--- Análise Descritiva da 'Growth Rate' ---")
estatisticas = df['Growth Rate'].describe()
print(estatisticas)
print("-" * 40)

# 2. Cálculo dos Limites para Detecção de Outliers (seu código original)
Q1 = df['Growth Rate'].quantile(0.25)
Q3 = df['Growth Rate'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

print("\n--- Limites Calculados para Outliers ---")
print(f"Limite Inferior (qualquer valor abaixo é outlier): {limite_inferior:.4f}")
print(f"Limite Superior (qualquer valor acima é outlier): {limite_superior:.4f}")
print("-" * 40)


# 3. Filtrar e Apresentar os Outliers (seu código original, com mais contexto)
outliers = df[(df['Growth Rate'] < limite_inferior) | (df['Growth Rate'] > limite_superior)]

print("\n--- Países com Taxa de Crescimento Considerada Outlier ---")
if outliers.empty:
    print("Nenhum outlier encontrado com base no método IQR.")
else:
    print(outliers[['Country', 'Growth Rate']].sort_values(by='Growth Rate', ascending=False))
print("-" * 40)
```

```python
# Faça um ranking dos continentes com base na sua taxa de crescimento (Growth Rate) média.

# Calcular a taxa de crescimento média para cada continente
taxa_media_continente = df.groupby('Continent')['Growth Rate'].mean().sort_values(ascending=False)

# Criar um DataFrame a partir do resultado para melhor visualização
df_ranking = taxa_media_continente.reset_index(name='Taxa de Crescimento Media')

print("Ranking de continentes por taxa de crescimento média:")
print(df_ranking)
```
